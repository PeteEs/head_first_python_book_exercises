```python
import os

import DBcm
```

```python
db_details = "CoachDB.sqlite3"
```

```python
FOLDER = "swimdata/"
```

```python
files = os.listdir(FOLDER)
files.remove(".DS_Store")
```

```python
SQL_SELECT  = """ 
    select * from events
    where distance = ? and stroke = ?
"""
```

```python
SQL_INSERT = """ 
    insert into events
        (distance, stroke)
    values
        (?, ?)
"""
```

```python
with DBcm.UseDatabase(db_details) as db:
    for file in files:
        _, _, distance, stroke = file.removesuffix(".txt").split("-")
        db.execute(SQL_SELECT, (distance, stroke))
        if db.fetchall():
            continue
        db.execute(SQL_INSERT, (distance, stroke))
```

```python
SQL_SELECT_ALL  = """ 
    select * from events
"""
```

```python
with DBcm.UseDatabase(db_details) as db:
    db.execute(SQL_SELECT_ALL)
    results = db.fetchall()

for result in results:
    print(result)
```
